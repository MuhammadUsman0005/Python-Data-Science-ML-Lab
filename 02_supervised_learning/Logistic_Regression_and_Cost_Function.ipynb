## Logistic_Regression_and_Cost_Function

```python
import numpy as np

# 1. Data Setup (4 inputs, 1=Spam, 0=Not Spam)
X = np.array([1.0, 2.0, 4.0, 5.0])
y = np.array([0.0, 0.0, 1.0, 1.0])

# 2. Parameters initialization
w = 0.0
b = 0.0
alpha = 0.2
iterations = 1000
m = len(X)

print("🚀 Training Shuru Ho Rahi Hai...\n")

# 3. Gradient Descent Loop
for i in range(iterations):
    # Step A: Sigmoid Prediction
    z = w * X + b
    f_x = 1 / (1 + np.exp(-z))
    
    # Step B: Cost Calculation (Log Loss)
    # epsilon use karte hain taake log(0) ka error na aaye
    eps = 1e-15
    f_x = np.clip(f_x, eps, 1 - eps)
    cost = - (1/m) * np.sum(y * np.log(f_x) + (1 - y) * np.log(1 - f_x))
    
    # Step C: Derivatives (Gradients)
    dj_dw = (1/m) * np.sum((f_x - y) * X)
    dj_db = (1/m) * np.sum(f_x - y)
    
    # Step D: Update Parameters
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
    
    # Progress check karein har 200 steps baad
    if i % 200 == 0:
        print(f"Step {i:4d}: Cost = {cost:.4f}, w = {w:.3f}, b = {b:.3f}")

print(f"\n🎯 Training Mukammal! Final Model: f(x) = Sigmoid({w:.2f}*x + {b:.2f})")

# Test karte hain: Agar kisi email mein 4 dafa 'Urgent' lafz aaye:
z_test = w * 4 + b
prob = 1 / (1 + np.exp(-z_test))
print(f"📧 4 Urgent words wali email ke Spam hone ka chance: {prob*100:.1f}%")
```
