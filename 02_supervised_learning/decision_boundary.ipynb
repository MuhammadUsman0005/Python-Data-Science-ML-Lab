## Decision Boundary in Logistic Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# 1. Fake Students Data (Maths, Coding)
# 🟢 Pass Students
pass_students = np.array([[8, 7], [6, 8], [9, 5], [7, 6]])
# 🔴 Fail Students
fail_students = np.array([[2, 3], [1, 5], [3, 2], [4, 4]])

# 2. Figure setup
plt.figure(figsize=(8, 6))

# 3. Data points plot karte hain
plt.scatter(pass_students[:, 0], pass_students[:, 1], color='green', s=100, label='Admitted (1)')
plt.scatter(fail_students[:, 0], fail_students[:, 1], color='red', s=100, label='Rejected (0)')

# 4. Decision Boundary Line khinchte hain (x1 + x2 = 10  ==>  x2 = 10 - x1)
x1_values = np.linspace(0, 10, 100)
x2_values = 10 - x1_values

plt.plot(x1_values, x2_values, color='blue', linewidth=3, linestyle='--', label='Decision Boundary (z=0)')

# 5. Graph ki decorations
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('Maths Score (x1)', fontsize=12, fontweight='bold')
plt.ylabel('Coding Score (x2)', fontsize=12, fontweight='bold')
plt.title('🎓 University Admission: Decision Boundary', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
