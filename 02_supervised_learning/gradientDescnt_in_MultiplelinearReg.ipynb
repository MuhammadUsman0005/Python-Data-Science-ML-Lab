## Gradient Descent for Multiple Linear Regression (MLR)

```python
import numpy as np

# 1. Data: 3 ghar hain, har ghar ke 2 features hain [Bedrooms, Size]
X = np.array([
    [1.0, 2.0],
    [2.0, 1.0],
    [3.0, 4.0]
])
y = np.array([5.0, 4.0, 11.0]) # Asli Qeemat (Target)
```

```python
# 2. Initial Settings
w = np.array([0.0, 0.0])  # Do features hain toh do weights
b = 0.0
alpha = 0.05              # Learning Rate
m = len(y)                # Total rows (3)
```

```python
# 3. Gradient Descent Loop (Hum 300 steps chalayenge)
for step in range(300):
    # Line 1: Vectorized Prediction (Matrix Dot Product)
    y_hat = np.dot(X, w) + b
    
    # Line 2: Ghalati nikaali (Vector of errors)
    error = y_hat - y
    
    # Line 3: Saare weights ki dhalan aik sath nikaali (X ka Transpose use kar ke)
    dj_dw = (1/m) * np.dot(X.T, error)
    
    # Line 4: Bias ki dhalan nikaali
    dj_db = (1/m) * np.sum(error)
    
    # Line 5: Weights Vector aur Bias ko update kiya
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
```

```python
print(f"🎯 MLR ne seekha -> Weights (w1, w2): {w.round(2)}, Bias (b): {b:.2f}")
```
