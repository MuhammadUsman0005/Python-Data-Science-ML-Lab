## Linear Regression Cost Funcion

```python
import numpy as np
import pandas as pd

# 1. Data tayyar karte hain
data = {
    'Pages': [100, 200, 300],
    'Asli_Price': [40, 120, 150]
}
df = pd.DataFrame(data)
```

```python
# 2. Hamara Cost Function (Andrew Ng style)
def compute_cost(X, y, w, b):
    m = len(X) # Total data points
    total_cost = 0
    
    for i in range(m):
        # Formula: y_hat = w*x + b
        y_hat = w * X[i] + b
        
        # Formula: (y_hat - y)^2
        error_squared = (y_hat - y[i]) ** 2
        
        # Summation
        total_cost += error_squared
        
    # Final Formula: 1/(2m) * total_cost
    final_cost = total_cost / (2 * m)
    return final_cost
```

```python
# 3. Test karte hain apne w aur b ke saath
X = df['Pages'].values
y = df['Asli_Price'].values

w_test = 0.5
b_test = 0

cost = compute_cost(X, y, w_test, b_test)
print(f"📉 Hamare model ka total cost (error) hai: {cost:.2f}")
```
