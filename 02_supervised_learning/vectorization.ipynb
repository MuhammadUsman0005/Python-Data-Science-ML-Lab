## Vectorization

```python
import numpy as np

# 1. Hamare paas 3 kitabein hain, aur har kitab ke 3 features hain:
# [Pages, Condition(1-10), Age(years)]
X = np.array([
    [100, 8, 2],   # Kitab 1
    [200, 9, 1],   # Kitab 2
    [300, 5, 5]    # Kitab 3
])
```

```python
# 2. AI ne training ke baad yeh Weights aur Bias dhoonde (Farz karein):
w = np.array([0.5, 10.0, -2.0]) 
b = 50.0

# Pages ka weight 0.5 (positive)
# Condition ka weight 10.0 (bohot positive)
# Age ka weight -2.0 (negative, kyunke purani kitab sasti hogi)
```

```python
# 3. LOOPS KE BAGHAIR - Aik hi jhatke mein sab ki prediction!
# np.dot har row ko w se multiply kar ke plus karega
predictions = np.dot(X, w) + b
```

```python
# 4. Results dekhte hain
for i in range(len(predictions)):
    print(f"📚 Kitab {i+1} ki predicted price: {predictions[i]} Rupees")
```
