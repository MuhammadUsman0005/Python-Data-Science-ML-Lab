## Gradient Descent In Linear Regression

```python
import numpy as np

# 1. Dataset (y = 2x hai, yaani perfect w=2 aur b=0 hona chahiye)
X = np.array([1.0, 2.0, 3.0, 4.0])
y = np.array([2.0, 4.0, 6.0, 8.0])
```

```python
# 2. Hyperparameters (Settings)
w = 0.0          # Shuruati w
b = 0.0          # Shuruati b
alpha = 0.01     # Qadam ka size (Learning Rate)
epochs = 500     # Loop kitni dafa chalega (Iterations)
m = len(X)       # Data points ki tadad
```

```python
# 3. Optimization Loop
for step in range(epochs):
    # Line 1: Saare points ke liye aik sath prediction ki
    y_hat = w * X + b
    
    # Line 2: w ki dhalan (Derivative) nikaali
    dj_dw = (1/m) * np.sum((y_hat - y) * X)
    
    # Line 3: b ki dhalan (Derivative) nikaali
    dj_db = (1/m) * np.sum(y_hat - y)
    
    # Line 4: w aur b ko aik sath update (badla) kiya
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
```

```python
# Final result check karte hain
print(f"🎯 AI ne khud seekha -> w: {w:.2f}, b: {b:.2f}")
```
