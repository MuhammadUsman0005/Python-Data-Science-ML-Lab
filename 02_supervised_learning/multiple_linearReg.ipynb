## Multiple Linear Regression

```python
import numpy as np

# 1. Data: 2 kitabein hain, har kitab ke 3 features hain: [Pages, Condition, Age]
X = np.array([
    [100, 8, 2],  # Kitab 1
    [200, 5, 6]   # Kitab 2
])
```

```python
# 2. Har feature ka apna weight vector (w) aur ek single bias (b)
w = np.array([0.5, 5.0, -1.5])  # w1=0.5, w2=5.0, w3=-1.5
b = 10.0
```

```python
# 3. Vectorization: Koi loop nahi, ek jhatke mein sab ki prediction!
# Matrix X aur Vector w ka Dot Product kiya aur b plus kiya
predictions = np.dot(X, w) + b
```

```python
# Results print karte hain
print(f"📚 Kitab 1 ki Qeemat: {predictions[0]} Rs")
print(f"📚 Kitab 2 ki Qeemat: {predictions[1]} Rs")
```
